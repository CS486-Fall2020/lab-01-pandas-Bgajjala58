```python
#Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
```

```python
df = pd.read_csv('ATP_World_Rankings.csv')
```

```python
#Inspect the dataset to determine the total winning percentage for each player.
#Taking the wins of each tennis player and converting them to percentage 
wl = df['Careers_Wins_Losses']
name = df['Tennis_Player']
Perc = []
for x in range(len(wl.index)):
    Player_numbers = wl[x].split('-')
    Player_total = int(Player_numbers[0]) + int (Player_numbers[1])
    Player_perc = int(Player_numbers[0])/Player_total
    Perc.append(round(Player_perc*100,2))
```

```python
Player_wining_percentagedf = {'Tennis_Player' : name,
                            'Careers_Wins_Percentage': Perc}
Player_wining_percentagedf= pd.DataFrame(Player_wining_percentagedf,columns =['Tennis_Player','Careers_Wins_Percentage'])
#Careers wins percentage 
Player_wining_percentagedf
```

```python
#Bar graph shows the diffrence between each player and how many weeks they are at number 1.
#This graph shows that the top are closely stacked in terms of weeks at number 1. After the top 5 there is a 
#Significant dip
rankdf = df.loc[df['Total_Weeks_at_No_1'] > 50]
rankdf.plot(x='Tennis_Player', fontsize=20, y='Total_Weeks_at_No_1',kind='bar',rot=0,figsize=(40,20),color=(0.2, 0.4, 0.6, 0.6))
```

```python
#Based on the scatter plot below we can see around 75 to 80 percent we see a spike in prize money.
#0 to 75 percent it is constant
relationdf = df.merge(Player_wining_percentagedf)
relationdf.plot('Careers_Wins_Percentage', 'Prize_Money', kind='scatter',rot=0,color=(0.2, 0.4, 0.6, 0.6))
```

```python
#Getting career total games per player 
wl = df['Careers_Wins_Losses']
name = df['Tennis_Player']
Perc2 = []
for x in range(len(wl.index)):
    Player_numbers = wl[x].split('-')
    Player_total = int(Player_numbers[0]) + int (Player_numbers[1])
    Perc2.append(round(Player_total,2))
```

```python
#The Histogram below shows that most players played around 875 to 950 games.
Player_game_totaldf = {'Tennis_Player' : name,
                            'Total_Games': Perc2}
Player_game_totaldf = pd.DataFrame(Player_game_totaldf,columns =['Tennis_Player','Total_Games'])
Player_game_totaldf.plot(x='Tennis_Player', fontsize=20, y='Total_Games',kind='hist',rot=0,figsize=(40,20),color=(0.2, 0.4, 0.6, 0.6))
```

```python
#To conclude each player plays around 800 to 950 games. Players around 75 to 80 percent winrate see a increase in pay
#from prize money. And from there the pay keeps increasing with Federer at top, with the second highest win rate and
#highest pay.
```
